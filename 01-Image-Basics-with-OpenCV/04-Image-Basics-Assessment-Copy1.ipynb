```python
vertices = np.array([[100,820], [500,400], [550,820]], dtype=np.int32)
points = vertices.reshape(-1,1,2)
cv2.polylines(img, [points], isClosed=True, color=(0,0,255), thickness=8)
plt.imshow(img)
```

```python
vertices = np.array([[100,820], [500,400], [550,820]], dtype=np.int32)
points = vertices.reshape(-1,1,2)
cv2.fillPoly(img, [points], (0,0,255))
plt.imshow(img)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import sys
sys.path.append('C:/Users/user/anaconda3/envs/python-cvcourse/lib/site-packages')

import cv2
import numpy as np


################
### Function ###
################

def draw_circle(event, x, y, flags, param):  #  callback과 연결, event는 클릭과 같은 동작, 좌표는 opencv가 알아서 받아줌, flags, param는 따로 필요한 경우 입력됨
    
    if event == cv2.EVENT_RBUTTONDOWN:  # 왼쪽 버튼 눌림 (눌리는 순간 event 발생)
        cv2.circle(img, (x, y), 50, (123,5,200), 5)
    

# namedWindow로 연결하고(윈도우 이름이 imshow의 것과 같아야 함)
cv2.namedWindow(winname='my_drawing')

# callback에 윈도우 이름과 함수를 씀 (함수와 윈도우가 연결되는 것)
cv2.setMouseCallback('my_drawing', draw_circle)


#################################
### Showing image with opencv ###
#################################

img = cv2.imread('../DATA/dog_backpack.jpg')

while True:
    cv2.imshow('my_drawing', img)
    
    if cv2.waitKey(1) & 0xff == 27:
        break
        
cv2.destroyllWindows()
```
