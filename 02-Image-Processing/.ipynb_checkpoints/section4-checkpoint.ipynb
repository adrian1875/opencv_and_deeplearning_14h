- RGB는 오래된 컬러 모델
- HSL(색조, 채도, 밝기)와 HSV(색조, 채도, 명조)가 새로운 컬러 모델로 1970년대에 등장함
- 위 두 가지는 인간이 실제 색을 인지하는 것과 밀접한 연관성이 있다
---
- RGB는 컬러 큐브 형태로 나타낼 수 있으며 RGB가 각각 0~255 사이의 값을 지닌다
- 큐브 안의 색을 보았을 때 R=1/5, G=3/5, B=4/5라면 실제로는 각각에 255를 곱한 값이 해당 컬러의 값이 된다 (해당 컬러가 0~255사이에서 각 색을 얼마의 비율로 가졌느냐를 보아 각 컬러의 레벨이 얼마나 높은가를 알 수 있다) 
---
- HSL은 색조, 채도, 밝기(Hue, Saturation, Lightness)를 나타내는 원기둥으로 표현함 
- 색조는 원기둥의 둘레, 채도는 원기둥의 반지름이다
- 원의 둘래는 초록, 노랑, 빨강 등과 같은 다른 색을 의미하고, 채도는 진함과 옅음을 나타내며, 밝기는 어둡고 밝음의 정도를 나타낸다
- 원통의 가장 윗 부분은 흰색이며 이는 최종 밝기는 모두 완벽한 흰색임을 의미한다 (반대로 바닥 부분은 완전한 검은색)
---
- HSV는 색조, 채도, 명도(Hue, Saturation, Value)를 나타내는 원기둥으로 표현
- 채도는 특정 색의 선명도에 대비해 얼마나 흰색에 가까운지를 나타냄
- 원기둥의 가장 윗 부분은 같은 색조를 띠며 색조와 채도 만을 정해주면 된다
- 원통의 정 중앙이 흰색, 즉 어떤 색조를 가지더라도 채도가 0이면 흰색이다

- 이미지 블랜딩하기
- opencv의 addWeighted 함수를 통해 이루어짐
- new_pixel = α * pixel_1 + β * pixel_2 + γ
- 두 이미지를 사용하여 이미지나 이미지 픽셀에 추가된 웨이트를 결합해서 새로운 픽셀에 결과를 나타냄
- 새로 블랜딩 된 이미지의 새로운 픽셀은 선택한 알파값에 첫 이미지의 픽셀값을 곱하고, 선택한 베타 값에 두 번째 이미지 픽셀값을 곱하고 이들을 더해서 구한다
- 추가로 감마 값을 더해주어도 됨
- 가중 합을 구하는 것
- 첫 이미지의 첫 픽셀이 도드라지는 것을 원하면 알파 값을 높게 입력하면 되고 두 번째 이미지가 덜 도드라져 보이게 만들고 싶다면 베타값을 낮추면 된다

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('../DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
img2.shape
```

```python
x_offset = 934 -600
y_offset = 1401 -600
```

```python
# 전경 이미지 ROI
rows, columns, channels = img2.shape

roi = img1[y_offset:1401, x_offset:934]  # 넘파이는 사용자가 보았을 때 세로축을 x축으로 간주한다, 반대로 가로 축을 y축으로 간주함
plt.imshow(roi)
```

```python
img2Gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

# opencv의 cvtColor로 gray 스케일로 변환을 하게 되면 단일 체널 이미지가 되는데 이걸 plt에서는 어느 컬러맵에 매칭해야 하는지 모른다
# 그래서 plt.imshow()에는 인자로 cmap이 존재하고, 이를 통해 출력 컬러맵을 설정할 수 있는데 보통 viridis 컬러맵으로 설정이 된다.
# 그래서 매개변수로 cmap='viridis'를 주면 매개변수 없이 default 값과 같은 이미지가 출력된다.
plt.imshow(img2Gray, cmap='gray')
```

```python
# 마스크로 사용할 것이기 때문에 gray로 변환된 부분 중 검은 색으로 된 부분이 필요해서 이미지를 반전하여 검은색 부분을 흰색으로 바꾸어 주야야 한다
# The function cv::bitwise_not calculates per-element bit-wise inversion of the input
# 비트 연산으로 반전한다
mask_inv = cv2.bitwise_not(img2Gray)

plt.imshow(mask_inv, cmap='gray')
```

```python
# gray는 단일 채널 이므로 다시 채널 3개를 추가해야 한다
print(mask_inv.shape)
```

```python
# 방법은 여러가지 임 (img2와 크기가 같도록 np.이라고 적어주고 255를 곱하는 방법이 있고, 아래와 같은 방법도 있음)
# Return a new array of given shape and type, filled with fill_value.
# 여기서는 img2의 사이즈에 맞는 흰 배경 이미지를 만든 것
white_background = np.full(img2.shape,255, dtype=np.uint8)
#print(white_background.shape)

# 비트 연산 or을 해줌 (bitwise OR is true if either of the two pixels is greater than zero)
# Two arrays when src1 and src2 have the same size
# param mask optional operation mask, 8-bit single channel array, that specifies elements of the output array to be changed
# mask된 영역을 기준으로 두 이미지를 or연산 하는 것
# 흰 배경 2개 위에 마스크를 씌운 것
# 이미지가 복잡하거나 색상이 1개 이상인 경우 bitwise를 사용하는 것이 좋음
bk = cv2.bitwise_or(white_background,white_background, mask=mask_inv)
#print(bk.shape)
#plt.imshow(bk)

#plt.imshow(cv2.absdiff(white_background, bk))
```

```python
plt.imshow(bk)
```

```python
# 이제 이걸 빨강으로 바꿀 것
# 이것이 특정 부분만 보이게 하는 mask의 역할아댜
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg)
```

```python
# 합성!!
final_roi = cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi

large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img

# final step
plt.imshow(large_img)
```
