```python
# 사용자 자정 도형 그리기
blank_img = np.zeros(shape=(640,640,3), dtype=np.int32)
plt.imshow(blank_img)
```

```python
# step1 꼭짓점 정하기
#    도형을 만들기 위해 그 꼭짓점을 지정하는 것 각 리스트는 x,y쌍이 된다
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
```

```python
vertices.shape
```

```python
# step2 꼭짓점 reshape
#    opencv는 3D정보가 필요하다
#    reshape(-1,1,2)를 가장 많이 사용한다 이렇게 하면 각 색상 체널마다 점이 생긴다
points = vertices.reshape(-1,1,2)
```

```python
points.shape
```

```python
vertices
```

```python
points
```

```python
# step3 그리기
#    원본 이미지, reshape한 꼭짓점 리스트(int32 or int64), 닫고싶다면 isClosed=True
cv2.polylines(blank_img, [points], isClosed=False, color=(255,250,10), thickness=4)
plt.imshow(blank_img)
```

```python
import cv2
import numpy as np


################
### Function ###
################

def draw_circle(event, x, y, flags, param):  #  callback과 연결, event는 클릭과 같은 동작, 좌표는 opencv가 알아서 받아줌, flags, param는 따로 필요한 경우 입력됨
    
    if event == cv2.EVENT_LBUTTONDOWN:  # 왼쪽 버튼 눌림 (눌리는 순간 event 발생)
        cv2.circle(img, (x, y), 50,(123,5,200),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 50,(0,0,240),-1)


# namedWindow로 연결하고(윈도우 이름이 imshow의 것과 같아야 함)
cv2.namedWindow(winname='my_drawing')

# callback에 윈도우 이름과 함수를 씀 (함수와 윈도우가 연결되는 것)
cv2.setMouseCallback('my_drawing', draw_circle)



#################################
### Showing image with opencv ###
#################################

img = np.zeros(shape=(512,512,3),dtype=np.int8)  # np.int8는 배경을 더 gray 처럼 보이게 해줌 완전 black을 원하면 안쓰면 된다 

while True:
    cv2.imshow('my_drawing', img)
    
    if cv2.waitKey(20) & 0xff == 27:
        break
        
cv2.destroyllWindows()
```

```python
import cv2
import numpy as np

# variables
# True while mouse button down, False while mouse button up
drawing = False
ix, iy = -1, -1

# function
def draw_rectangle(event, x, y, flags, param):
      
    global ix, iy, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y  # 마우스가 눌린 위치를 기록
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy), (x,y), (255,255,250), -1)  # 여기서 x,y는 버튼에서 손을 땐 시점(EVENT_LBUTTONDOW은 눌린 순간에 event가 발생)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy), (x,y), (255,255,250), -1)

# showing the image
img = np.zeros((512,512,3))  # black

cv2.namedWindow(winname= 'my_drawing')

cv2.setMouseCallback('my_drawing', draw_rectangle)

while True:
    cv2.imshow('my_drawing', img)
    
    # check for esc
    if cv2.waitKey(1) & 0xff == 27:
        break
        
cv2.destroyllWindows()
```
